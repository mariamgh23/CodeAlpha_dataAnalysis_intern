```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
```

```python
Car_Price = []
Location = []
Brand = []
Milage = []
year = []
Type = []
```

```python
base_url = "https://www.dubizzle.com.eg/en/vehicles/cars-for-sale/?page={}"
page_num = 1

while True:
    response = requests.get(base_url.format(page_num))
    soup = BeautifulSoup(response.content, "lxml")
    
    # Extract data on current page
    price = soup.find_all("span", {"class": "ddc1b288"})
    loc = soup.find_all("span", {"class": "f7d5e47e"})
    brand = soup.find_all("div", {"class": "_34bc0d5f c6f55f20 _81d9cd00"})
    block = soup.find_all("span", class_="_3e1113f0")
    
    if not price:  # No more listings
        break

    for i in range(len(price)):
        Car_Price.append(price[i].text if i < len(price) else "N/A")
        Location.append(loc[i].text if i < len(loc) else "N/A")
        Brand.append(brand[i].text if i < len(brand) else "N/A")

    for i in range(0, len(block), 3):
        group = block[i:i+3]
        if len(group) == 3:
            year.append(group[0].text.strip())
            Milage.append(group[1].text.strip())
            Type.append(group[2].text.strip())
    
    print(f"Scraped page {page_num}")
    page_num += 1
```

```python
file=[year,Brand,Car_Price,Location,Milage,Type]
exported=zip_longest(*file)
```

```python
with open("Car_Egypt_all.csv",mode="w",newline="",encoding="utf-8-sig") as file:
    writer=csv.writer(file)
    writer.writerow(["year","Brand","Car_Price","Location","Mileage","Type"])
    for i in exported:
        writer.writerow([i[0],i[1],i[2],i[3],i[4],i[5]])
```
